# Linear Regression in Python


```python
import random

def simple_linear_regression(data, steps=1000, alpha=0.02):

    b0 = random.randint(-10,10)  # initialize with random value
    b1 = random.randint(-10,10)  # initialize with random value
    n = len(data)

    for step in range(steps):

        # adjust intercept
        # b0 = b0 – alpha * 2 / n * sum(-(yi - (b0 + b1 * xi))
        sum0 = 0
        for d in data:
            sum0 += ((b0 + b1 * d[0]) - d[1])
        b0 = b0 - (alpha * 2/n * sum0)

        # adjust slope
        # b1 = b1 – alpha * 2 / n * sum(- xi * (yi - (b0 + b1 * xi))
        sum1 = 0
        for d in data:
            sum1 += d[0] * ((b0 + b1 * d[0]) - d[1])
        b1 = b1 - (alpha * 2/n * sum1)

        sse = 0
        for d in data:
            sse += (d[1] - (b0 + b1 * d[0]))**2
        mse = sse/n
        print("b0={}, b1={}".format(b0, b1))
        print("Step {}, MSE= {}".format(step, mse))

    return (b0, b1)

```

## Dummy Example

```python

data_dummy = [
    (1, 150),
    (2, 250),
    (3, 350),
    (4, 450),
    (5, 550),
    (6, 650),
    (7, 750)
]


simple_linear_regression(data_dummy, steps=1000)
```

## Sales example

```python

data_sales = [
    (1.7, 368),
    (1.5, 340),
    (2.8, 665),
    (5.0, 954),
    (1.3, 331),
    (2.2, 556),
    (1.3, 376)
]

simple_linear_regression(data_sales)
```
